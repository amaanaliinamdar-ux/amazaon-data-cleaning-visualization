```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', "_")
```

```python
df.head(6)
```

```python
df = df.drop('product_title',axis=1)
```

```python
df.drop(columns='technical_details_dict',axis=1)
df.head(2)
```

```python
df.sample(3)
```

```python
df['total_rating'] = df['total_rating'].str.strip('()')
```

```python
df.sample(3)
```

```python
df['total_rating'] = df['total_rating'].str.replace(',','').astype(int)
```

```python
df.sample(3)
```

```python
df['price'] = df['price'].str.replace(',','')
df['price'] = df['price'].str.strip('.').astype(int)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
df.sample(3)
```

```python
df.sample(3)
```

```python
df['mrp'] = df['mrp'].str.replace(',','')
df['mrp'] = df['mrp'].str.strip('₹')
df.sample(3)
```

```python
df = df.drop('product_details_dict',axis=1)
df = df.drop(columns='special_features',axis=1)
df = df.drop(columns='graphics_card',axis=1)
df.sample(3)
```

```python
df['offer'] = df['offer'].str.strip('-%')
df.sample(3)
```

```python
df['colour'].value_counts()
```

```python
df.loc[df['colour'].str.contains(r'\d|\|', na=False), 'colour'] = pd.NA
```

```python
df.sample(5)
```

```python
df['screen_size'].value_counts()
```

```python
df.head(2)
```

```python
df['screen_size_num'] = df['screen_size'].str.extract(r'([\d.]+)').astype(float)


cm_mask = df['screen_size'].str.contains('Centimetres', case=False, na=False)


df.loc[cm_mask, 'screen_size_num'] = df.loc[cm_mask, 'screen_size_num'] / 2.54


df.rename(columns={'screen_size_num': 'screen_size_inches'}, inplace=True)
```

```python
df.head(4)
```

```python
df = df.drop('screen_size',axis=1)
df.sample(3)
```

```python
df['hard_disk_size'] = df['hard_disk_size'].str.replace('GB', '', case=False, regex=False).str.strip()


df['hard_disk_size_gb'] = df['hard_disk_size'].str.extract(r'([\d.]+)').astype(float)


tb_mask = df['hard_disk_size'].str.contains('T', case=False, na=False)


df.loc[tb_mask, 'hard_disk_size_gb'] *= 1024
```

```python
df.head(3)
```

```python
df = df.drop('hard_disk_size',axis=1)
df.sample(3)
```

```python
df.sample(5)
```

```python
df.rename(columns={'ram': 'ram_in_gb'}, inplace=True)
```

```python
df.sample(3)
```

```python
df.columns
```

```python
df.dropna(subset=['price', 'rating'], inplace=True)
```

```python
df.sample(3)
```

```python
df['rating'] = df['rating'].fillna(df['rating'].median())
df['offer'] = df['offer'].fillna(0)
```

df.sample(3)

```python
df.head(3)
```

```python
df['brand'] = df['brand'].fillna('Unknown',)
df['colour'] = df['colour'].fillna('Unknown',)
df['model_name'] = df['model_name'].fillna('Unknown',)
df['cpu_model'] = df['cpu_model'].fillna('Unknown',)
df['os'] = df['os'].fillna('Unknown',)
df['model_name'] = df['model_name'].fillna('Unknown',)
df['special_features'] = df['model_name'].fillna('Unknown',)
df['graphics_card'] = df['model_name'].fillna('Unknown',)
```

```python
df.head(3)
```

```python
df['about'] = df['about'].fillna('',)
```

```python
df.head(3)
```

```python
import seaborn as sns

sns.histplot(df['price'], bins=30, color='skyblue', kde=True)
plt.title('Price Distribution (Seaborn)')
plt.show()
```

```python
sns.scatterplot(x='price', y='rating', data=df, alpha=0.6, color='green')
plt.title('Price vs Rating (Seaborn)')
plt.show()
```

```python
sns.boxplot(x='brand', y='price', data=df)
plt.xticks(rotation=45)
plt.title('Price by Brand (Seaborn)')
plt.show()
```

```python
sns.countplot(x='brand', data=df)
plt.xticks(rotation=45)
plt.title('Number of Products by Brand (Seaborn)')
plt.show()
```
